```python
import math as m
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel('sonar.all-data.xlsx')
df
```

```python
type_dummies=pd.get_dummies(df['type'])
```

```python
df_DCT=pd.read_excel('Y축.xlsx')
df_DCT
```

```python
for p in range(0, 207):
  r=int(p)
  l=[]
  for k in range(0, 61):
    T=0
    for n in range(0, 60):
      c=int(n)
      s=int(k)
      b=float(df.iloc[r, c])
      a=b*m.cos((k*m.pi*(2*c+1))/120)
      T=a+T
    l.append(T)
  df_DCT[r]=l
df_DCT
```

```python
fig,ax=plt.subplots()
for p in range (0, 96):
  q=int(p)
  ax=sns.scatterplot(data=df_DCT,x='frequency',y=q, color='blue', alpha=0.3)
for r in range (96, 207):
  s=int(r)
  ax=sns.scatterplot(data=df_DCT,x='frequency',y=s, color='red', alpha=0.3)
```
